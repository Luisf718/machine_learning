```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset_salario_minimo = pd.read_csv("salario_minimo_mexico.csv")
print(dataset_salario_minimo)
```

```python
print(dataset_salario_minimo)
```

```python
x = dataset_salario_minimo.iloc[:, 0].values
y = dataset_salario_minimo.iloc[:, 3].values
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#regressor = LinearRegression()
regressor = DecisionTreeRegressor()
x_train= x_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)#
regressor.fit(x_train, y_train)
```

```python
#Graficamos el entrenamiento
viz_train = plt
viz_train.scatter(x_train, y_train, color = "blue")
viz_train.plot(x_train, regressor.predict(x_train), color = "black")
viz_train.title("Salario minimo mexico")
viz_train.xlabel("Año")
viz_train.ylabel("Salario minimo")
viz_train.show()
```

```python
#Graficamos el test
viz_train = plt
viz_train.scatter(x_test, y_test, color = "blue")
viz_train.plot(x_train, regressor.predict(x_train), color = "black")
viz_train.title("salario minimo")
viz_train.xlabel("Año")
viz_train.ylabel("Salario minimo")
viz_train.show()
```

```python
regressor.score(x_test, y_test)
```
